## 1. Counting birds

### <u> Counting birds part 1 </u>

Every year there is a day where people are encouraged to count the birds in their garden. Joseph tries to participate every year. Below you see the data of Joseph's observations. Let's make a nice plot of the data, also include the total amount of birds spotted.

```python
# Sample data
data = {
    'year': ['2020', '2021', '2022', '2023', '2024'],
    'Pigeons': [33, 52, 26, 40, 38],
    'Sparrows': [54, 48, 66, 59, 61],
    'others': [104, 98, 88, 92, 101]
}
```

### Inspecting the data

Load and inspect the data.

```python
df = pd.DataFrame(data)
df
```

### Plotting the pigeon count

Plot the number of counted pigeons over the year, following these specifications:

* Line graph
* Size of the figure: (10,6)
* Full blue line, data point marked with a circle, linewidth 2, marker size 8
* Provide the necessary labels in order to interpret the graph and show a legend
* Provide a meaningfull title
* Show a grid

```python
fig = plt.figure(figsize=(10, 6))

plt.plot(df["year"], df["Pigeons"], linestyle="-", color="Blue", marker="o", linewidth=2, markersize=8, label='Pigeon')
plt.xlabel('Year')
plt.ylabel('Amount')
plt.legend()


plt.title("Pigeon Count")
plt.grid()
plt.show()





```

### Plotting the sparrow count

Plot the number of sparrows Joseph counted over the years, adhere to the specifications listed below:

* Line graph
* linestyle: dotted
* color of the graph: Green
* marker: a triangle pointing to the right
* the edge of the markers should be red with width 1
* Corresponding labels and titles

Ensure the plot of the Pigeon count is incorporated as well.


```python
fig = plt.figure(figsize=(10,6))

plt.plot(df["year"], df["Sparrows"], linestyle="--", color="green", marker=">", markeredgecolor="red", markersize=10, label="Sparrow")

plt.xlabel("Year")
plt.ylabel("Count")
plt.legend()

plt.title("Sparrows")

plt.grid()

plt.show()


```

### Plotting the other birds

Add a graph line for all other birds counted. You can use a styling of your choice.

```python
fig = plt.figure(figsize=(10,6))

plt.plot(df["year"], df["Sparrows"], linestyle="--", color="green", marker=">", markeredgecolor="red", markersize=10, label="Sparrow")

plt.plot(df["year"], df["others"], linestyle="--", color="red", marker=">", markeredgecolor="black", markersize=10, label="others")

plt.xlabel("Year")
plt.ylabel("Count")
plt.legend()

plt.title("Sparrows + Others")

plt.grid()

plt.show()
```

### <u> Counting birds part 3 </u>

### Bar chart of total
Display the data  using a bar chart (total) and a stacked bar chart (pigeons, sparrows, others). Ensure the total is displayed on the right side and the stacked bar on the left.
Make it informative (and pretty).

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

fig, ax = plt.subplots(figsize=(12, 6))
bar_width = 0.35

x_pos = np.arange(len(df["year"]))

others = df["others"].values
sparrows = df["Sparrows"].values
pigeons = df["Pigeons"].values
totals = df["total"].values

# Stacked bar (links)
ax.bar(x_pos - bar_width / 2, others, bar_width, color="cornflowerblue", label="Others")
ax.bar(x_pos - bar_width / 2, sparrows, bar_width, bottom=others, color="tomato", label="Sparrows")
ax.bar(x_pos - bar_width / 2, pigeons, bar_width, bottom=others + sparrows, color="seagreen", label="Pigeons")

# Total bar (rechts)
ax.bar(x_pos + bar_width / 2, totals, bar_width, color="purple", label="Total")

ax.set_xlabel("Year")
ax.set_ylabel("Bird Count")
ax.set_title("Bird Count per Year: Stacked vs Total")
ax.set_xticks(x_pos)
ax.set_xticklabels(df["year"])
ax.legend(loc="upper left")

plt.tight_layout()
plt.show()
```

### <u> Counting birds part 4 </u>

### Grouped bar chart

Display the birds data using a grouped bar chart. Place the count of the pigeons and sparrows on the left of the x-tick, others and total on the right. Ensure the bars do not overlap.

```python
plt.figure(figsize=(10,6))

bar_width = 0.2

x = range(len(df["year"]))

others = plt.bar([i + bar_width for i in x], df["others"], bar_width, color='blue', label='Others', align="center")
Sparrows = plt.bar(x, df["Sparrows"], bar_width, color='Red', label='Sparrows', align="center")
Pigeons = plt.bar([i - bar_width for i in x], df["Pigeons"], bar_width, color='Yellow', label='Pigeons', align="center")
Total = plt.bar([i + (2* bar_width) for i in x], df["total"], bar_width, color='Green', label='Total', align="center")

plt.bar_label(others)
plt.bar_label(Sparrows)
plt.bar_label(Pigeons)
plt.bar_label(Total)

plt.xticks(x, df["year"])

plt.xlabel("Year")
plt.ylabel("Count")
```
